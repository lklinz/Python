```python
'''creating generic encrytion'''
def general_encrypt(plaintext_file,ciphertext_file,lower_case_dict,upper_case_dict):
    with open(f'{plaintext_file}','r') as read_file:
        with open(f'{ciphertext_file}','w') as write_file:
            for line in read_file:
                line = line.strip()
                encrypted_line = ''
                for letter in line:
                    if letter in lower_case_dict.keys():
                        letter = lower_case_dict[letter]
                    elif letter in upper_case_dict.keys():
                        letter = upper_case_dict[letter]
                    else:
                        letter = letter
                    encrypted_line += letter
                write_file.writelines(encrypted_line + '\n')
```

```python
def rot13_encrypt(plaintext_file,ciphertext_file):
    """ creating lower case and upper case alphabet for creating cypher because python is case sensitive. This wil avoid key error occur when encrypting the text"""
    
    lower_case_alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    lower_case_dict = {}
    upper_case_dict = {}   
    for i in range(13):
        lower_case_dict[lower_case_alphabet[i]] = lower_case_alphabet[i+13]
        lower_case_dict[lower_case_alphabet[i+13]] = lower_case_alphabet[i] 
        upper_case_dict[lower_case_alphabet[i].upper()] = lower_case_alphabet[i+13].upper()
        upper_case_dict[lower_case_alphabet[i+13].upper()] = lower_case_alphabet[i].upper()
    
    '''encrypting the plaintext file'''

    general_encrypt(plaintext_file,ciphertext_file,lower_case_dict,upper_case_dict)
    
rot13_encrypt('liquid.txt', 'rot13_cipher.txt')
```

```python
'''Creating function for decrypting encrypted text file utilizing the function for encrypting'''

def rot13_decrypted(ciphertext_file, plaintext_file):
    rot13_encrypt(ciphertext_file,plaintext_file)
rot13_decrypted('rot13_cipher.txt','liquid_decrypted_rot13.txt')
```

```python
def keyword_encrypt(plaintext_file, ciphertext_file, keyword):
    lower_case_alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    
    '''checking the validity of the keyword'''

    for letter in keyword:
        if keyword.count(letter) >1:
            sys.exit()

    '''creating the dictionaries for lower case and upper case letters'''

    lower_case_dict = {}
    upper_case_dict = {}
    for i in range(len(keyword)):
        lower_case_dict[lower_case_alphabet[i]] = keyword[i]
        upper_case_dict[lower_case_alphabet[i].upper()] = keyword[i].upper()

    a = len(keyword)
    for character in lower_case_alphabet:
        if character in keyword:
            continue
        else:
            lower_case_dict[lower_case_alphabet[a]] = character
            upper_case_dict[lower_case_alphabet[a].upper()] = character.upper()
        a = a+1

    '''calling the function to encrypt the plaintext file'''

    general_encrypt(plaintext_file, ciphertext_file,lower_case_dict,upper_case_dict)

keyword_encrypt('liquid.txt','keyword_cipher.txt','gojira')
```

```python
def atbash_encrypt(plaintext_file, ciphertext_file):
    lower_case_alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    cipher_upper = dict()
    cipher_lower = dict()
    # here we made a upper and lower dict, down here we used range function to incrament all letters with their corresponding cipher
    for i in range(len(lower_case_alphabet)):
        cipher_lower[lower_case_alphabet[i]] = lower_case_alphabet[-i-1]
    for letter in cipher_lower.keys():
        cipher_upper[letter.upper()] = cipher_lower[letter].upper()
    
    general_encrypt(plaintext_file,ciphertext_file,cipher_lower,cipher_upper)
                        
def atbash_decrypt(ciphertext_file, plaintext_file):
    atbash_encrypt(ciphertext_file, plaintext_file)
```
