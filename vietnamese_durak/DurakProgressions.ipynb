## Rules of Tấn (Durak):
Tấn (also known as Durak) is a Vietnamese card game, containing 2 to 4 players (in this game:2). At the beginning of the game, one of the players becomes the dealer and deals 8 cards to each player. A trump suit is then selected by pulling a random card from the deck and placing it face up on the table. (The trump suit can be used to defend against all other suits regardless of the power of the attacking cards. When using a triumph suit card to defend another triumph suit card, the player needs to follow the normal rules of the game). The remaining cards are then placed on the trump card, face down, leaving the suit of the trump visible.   

Each player takes turns to attack the player next to them (the first attacker chooses whether the game goes clockwise or counter clockwise). The attacked player then becomes the defender. After the attacker finishes their attack, any non-defending players can then choose to attack. The defender will continue to defend until either the attackers give up or the defender folds. 

The defender defends attacks by defending with a higher card of the attacking suit, or any card of the trump suit if applicable. If the defending player was successful, all played cards are removed from play. If the defender cannot defend or chooses not to, they are required to add all played cards to their hand. After a round terminates, the first attacker draws till they have 8 cards in their hand, at which point the next attacker draws till 8. This continues until all attackers have drawn till 8 or until the deck runs out of cards.  

The winner can be determined when: 

* The defender wins if the deck has no remaining cards and the defender successfully defends until they have no remaining cards.  

* If there are still cards remaining in the deck and the defender runs out of cards, but the attackers are unable to continue attacking, the defender wins.  

Otherwise, the attackers can continue attacking and then the defender must draw to 8 to continue defending.  

If a winner has been determined, the game will continue until there is only a single player with cards in their hand.
